```python
import pandas as pd
import numpy as np

# Assuming 'data' is your DataFrame

# Calculate the distribution of non-NaN values in 'a/w' and 'd/w'
a_w_distribution = data['a/w'].value_counts(normalize=True)
d_w_distribution = data['d/w'].value_counts(normalize=True)

# Fill NaN values in 'a/w' proportionally
data['a/w'] = data['a/w'].apply(lambda x: np.random.choice(a_w_distribution.index, p=a_w_distribution.values) if pd.isna(x) else x)

# Fill NaN values in 'd/w' proportionally
data['d/w'] = data['d/w'].apply(lambda x: np.random.choice(d_w_distribution.index, p=d_w_distribution.values) if pd.isna(x) else x)
```
